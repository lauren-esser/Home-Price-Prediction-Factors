```python
#Create IQR function for easier removal
def IQR_removal(col, test):
    Q1 = test[col].quantile(0.25)
    Q3 = test[col].quantile(0.75)
    IQR = Q3-Q1
    print(IQR)

    num_rows = len(test)
    test = test[~((test[col] < (Q1 - 1.5 * IQR)) |(test[col] > (Q3 + 1.5 * IQR)))]
    num_outliers = num_rows - len(test)
    print(test.shape)
    print(f"Number of outliers removed from {col}: {num_outliers}")
    return test 
    
```

I am using the observations from the describe performed previously to take note of which columns need outlier removal. Below is copied and pasted from before.

```python
def color_negative(val):
    color = "red" if val < 0 else 'black'
    return 'color: %s' % color
```

```python
coeffs = model.params.sort_values(ascending=False)
#if you increase by one thats the increase in selling point. 
coeffs.sort_index().to_frame('Coefficients').style.applymap(color_negative)
```

**Summary**:
>The coefficients show the relationship between each independent variable and the dependent variable of price. The higher the number the greater the positive relationship between the two. Since I performed a log transformation on bedrooms, bathrooms, sqft_lot, floors, lat, sqft_above, yr_built, sqft_living15, grade we will need to divide these columns by 100 to see how the 1% increase or decrease in these independent columns impact price. Some observations are listed below:
- **Bedroom** 33,444/100 = 334.44. A 1% increase of bedrooms increases the price by 334.44 dollars. Therefore adding more bedrooms to your house increases value.
- **Bathroom** A 1% increase in bathrooms causes a 374.88 dollar increase in price. 
- **Grade** Grade is the column that caused the largest increase to the dependent variable of price. This shows that houses built with better materials and are on the higher grade end will sell for much higher prices than lower ones. 2,635.13 dollars for every 1% increase. 
- **Waterfront** Waterfront was an item that I did not log transform. Thus living on waterfront increases the price of the house on average by 224995 dollars. This is usually out of the homeowners control, but something to consider when buying houses. 
- **Sqft_above** 1655.68 increase in price per 1% sqft_above. Sqft_above is the square footage of the house apart from the basement.
- **Sqft_living15** 802.39 increase in price per 1% sqft_living15. Sqft_living15 is define as Living room area in 2015.
- **View** View was not log transformed therefore increasing how many times the house was viewed before selling increases the homes price by 45796 dollars.
